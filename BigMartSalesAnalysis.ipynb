```python
m = ['Outlet_Size', 'Outlet_Type', 'Outlet_Location_Type', 'Item_Fat_Content', 'Item_Type']
X = pd.get_dummies(bigmart_train[m])
bigmart_train = pd.concat([bigmart_train, X], axis = 1)
```

```python
bigmart_test = pd.read_csv("/content/Test.csv")
```

```python
bigmart_test.drop(labels = ['Outlet_Establishment_Year'], inplace = True, axis = 1)
```

```python
k = ['Outlet_Size', 'Outlet_Type', 'Outlet_Location_Type', 'Item_Fat_Content', 'Item_Type']
X1 = pd.get_dummies(bigmart_test[k])
bigmart_test = pd.concat([bigmart_test, X1], axis = 1)
```

```python
bigmart_test['Outlet_Size'] = bigmart_test['Outlet_Size'].map({"Small":1, "Medium":2, "High":3})
```

```python
bigmart_test['Item_Weight'] = bigmart_test['Item_Weight'].fillna(bigmart_test['Item_Weight'].mean())
```

```python
bigmart_test['Outlet_Size'] = bigmart_test['Outlet_Size'].fillna(bigmart_test['Outlet_Size'].median())
```

```python
bigmart_train.drop(labels = ["Outlet_Size", "Outlet_Location_Type", "Outlet_Type", "Item_Fat_Content", "Item_Type", "Item_Identifier", "Outlet_Identifier"], axis = 1, inplace = True)
```

```python
bigmart_test.drop(labels = ["Outlet_Size", "Outlet_Location_Type", "Outlet_Type", "Item_Fat_Content", "Item_Type", "Item_Identifier", "Outlet_Identifier"], axis = 1, inplace = True)
```

```python
X_train = bigmart_train.drop(labels = ["Item_Outlet_Sales"], axis = 1)
y_train = bigmart_train["Item_Outlet_Sales"]
```

```python
X_train.shape, y_train.shape
```

```python
bigmart_train.head()
```

```python
y_train.head()
```

```python
from sklearn import preprocessing

x = X_train.values
testt = bigmart_test.values
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled_train = min_max_scaler.fit_transform(x)
x_scaled_test = min_max_scaler.fit_transform(testt)
df_train = pd.DataFrame(x_scaled_train)
df_test = pd.DataFrame(x_scaled_test)
```

```python
df_train.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_train, y_train, test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
```

```python
plt.scatter(y_test, pred)
plt.show()
```

```python
sns.distplot((y_test-pred), bins = 55)
plt.show()
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, pred))
print('MSE:', metrics.mean_squared_error(y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, pred)))
```
